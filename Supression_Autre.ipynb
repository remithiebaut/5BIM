# Récupération de la route des images que l'on veut supprimer

```python
current_dir = os.getcwd()
path_Donnee_DP_moitié_Autre = os.path.join(current_dir,'Donnee_DP_réduite')
```

# Boucle de supression d'une image sur deux

```python

for d in os.listdir(path_Donnee_DP_moitié_Autre):
    for dossier in os.listdir(os.path.join(path_Donnee_DP_moitié_Autre,d)):
        i = 0
        for classe in os.listdir(os.path.join(path_Donnee_DP_moitié_Autre,d,dossier)):
            if dossier == 'Autre' : 
                i = i+1
                if i%2 :
                    print(os.path.join(path_Donnee_DP_moitié_Autre,d,dossier,classe))
                    os.remove(os.path.join(path_Donnee_DP_moitié_Autre,d,dossier,classe))                
                    
```
